```python
import numpy as np  # Ÿàÿßÿ±ÿØ ⁄©ÿ±ÿØŸÜ ⁄©ÿ™ÿßÿ®ÿÆÿßŸÜŸá numpy ÿ®ÿ±ÿß€å ⁄©ÿßÿ± ÿ®ÿß ÿ¢ÿ±ÿß€åŸá‚ÄåŸáÿß€å ÿπÿØÿØ€å
import matplotlib.pyplot as plt  # Ÿàÿßÿ±ÿØ ⁄©ÿ±ÿØŸÜ ⁄©ÿ™ÿßÿ®ÿÆÿßŸÜŸá matplotlib ÿ®ÿ±ÿß€å ÿ±ÿ≥ŸÖ ŸÜŸÖŸàÿØÿßÿ±
from sklearn.linear_model import LinearRegression  # Ÿàÿßÿ±ÿØ ⁄©ÿ±ÿØŸÜ ŸÖÿØŸÑ ÿ±⁄Øÿ±ÿ≥€åŸàŸÜ ÿÆÿ∑€å ÿßÿ≤ sklearn

# ÿØÿßÿØŸá‚ÄåŸáÿß: ÿ≥ÿßÿπÿ™ ŸÖÿ∑ÿßŸÑÿπŸá Ÿà ŸÜŸÖÿ±Ÿá
hours = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)  # ÿ¢ÿ±ÿß€åŸá‚Äåÿß€å ÿßÿ≤ ÿ≥ÿßÿπÿßÿ™ ŸÖÿ∑ÿßŸÑÿπŸá (1 ÿ™ÿß 5)ÿå ÿ™ÿ®ÿØ€åŸÑ ÿ®Ÿá ÿ¢ÿ±ÿß€åŸá ÿØŸà ÿ®ÿπÿØ€å
scores = np.array([50, 60, 70, 80, 90])  # ÿ¢ÿ±ÿß€åŸá‚Äåÿß€å ÿßÿ≤ ŸÜŸÖÿ±ÿßÿ™ ŸÖÿ±ÿ®Ÿàÿ∑ ÿ®Ÿá Ÿáÿ± ÿ≥ÿßÿπÿ™ ŸÖÿ∑ÿßŸÑÿπŸá

# ÿ≥ÿßÿÆÿ™ ŸÖÿØŸÑ Ÿà ÿ¢ŸÖŸàÿ≤ÿ¥
model = LinearRegression()  # ÿ≥ÿßÿÆÿ™ €å⁄© ŸÖÿØŸÑ ÿ±⁄Øÿ±ÿ≥€åŸàŸÜ ÿÆÿ∑€å
model.fit(hours, scores)  # ÿ¢ŸÖŸàÿ≤ÿ¥ ŸÖÿØŸÑ ÿ®ÿß ÿØÿßÿØŸá‚ÄåŸáÿß (ÿ≥ÿßÿπÿßÿ™ ŸÖÿ∑ÿßŸÑÿπŸá Ÿà ŸÜŸÖÿ±ÿßÿ™)

# Ÿæ€åÿ¥‚Äåÿ®€åŸÜ€å ÿ®ÿ±ÿß€å €å⁄© ÿ≥ÿßÿπÿ™ ÿ¨ÿØ€åÿØ (ŸÖÿ´ŸÑÿß 6 ÿ≥ÿßÿπÿ™)
new_hour = np.array([[6]])  # Ÿàÿ±ŸàÿØ€å ÿ¨ÿØ€åÿØ ÿ®ÿ±ÿß€å Ÿæ€åÿ¥‚Äåÿ®€åŸÜ€å (€∂ ÿ≥ÿßÿπÿ™ ŸÖÿ∑ÿßŸÑÿπŸá)
predicted_score = model.predict(new_hour)  # Ÿæ€åÿ¥‚Äåÿ®€åŸÜ€å ŸÜŸÖÿ±Ÿá ÿ®ÿ±ÿß€å €∂ ÿ≥ÿßÿπÿ™ ŸÖÿ∑ÿßŸÑÿπŸá
print(f"If you study for 6 hours, your grade will be approximately: {predicted_score[0]:.1f}")  # ŸÜŸÖÿß€åÿ¥ ŸÜÿ™€åÿ¨Ÿá: ÿß⁄Øÿ± €∂ ÿ≥ÿßÿπÿ™ ÿØÿ±ÿ≥ ÿ®ÿÆŸàŸÜ€å ŸÜŸÖÿ±Ÿá‚Äåÿßÿ™ ÿ≠ÿØŸàÿØÿßŸã ⁄ÜŸÜÿØŸá

# ÿ±ÿ≥ŸÖ ŸÜŸÖŸàÿØÿßÿ±
plt.scatter(hours, scores, color='blue', label='Real Data')  # ÿ±ÿ≥ŸÖ ÿØÿßÿØŸá‚ÄåŸáÿß€å ŸàÿßŸÇÿπ€å ÿ®Ÿá ÿµŸàÿ±ÿ™ ŸÜŸÇÿ∑Ÿá‚ÄåŸáÿß€å ÿ¢ÿ®€å
plt.plot(hours, model.predict(hours), color='red', label='Learned line')  # ÿ±ÿ≥ŸÖ ÿÆÿ∑ €åÿßÿØ⁄Øÿ±ŸÅÿ™Ÿá‚Äåÿ¥ÿØŸá ÿ™Ÿàÿ≥ÿ∑ ŸÖÿØŸÑ ÿ®ÿß ÿ±ŸÜ⁄Ø ŸÇÿ±ŸÖÿ≤
plt.scatter(new_hour, predicted_score, color='green', s=100, label='New guess')  # ÿ±ÿ≥ŸÖ Ÿæ€åÿ¥‚Äåÿ®€åŸÜ€å ÿ¨ÿØ€åÿØ (ŸÜŸÇÿ∑Ÿá ÿ≥ÿ®ÿ≤ ÿ®ÿ≤ÿ±⁄Ø)
plt.xlabel('Study hours')  # ÿ®ÿ±⁄Üÿ≥ÿ® ŸÖÿ≠Ÿàÿ± ÿßŸÅŸÇ€å: ÿ≥ÿßÿπÿ™ ŸÖÿ∑ÿßŸÑÿπŸá
plt.ylabel('Score')  # ÿ®ÿ±⁄Üÿ≥ÿ® ŸÖÿ≠Ÿàÿ± ÿπŸÖŸàÿØ€å: ŸÜŸÖÿ±Ÿá
plt.title('Predicting scores with linear regression')  # ÿπŸÜŸàÿßŸÜ ŸÜŸÖŸàÿØÿßÿ±: Ÿæ€åÿ¥‚Äåÿ®€åŸÜ€å ŸÜŸÖÿ±Ÿá ÿ®ÿß ÿ±⁄Øÿ±ÿ≥€åŸàŸÜ ÿÆÿ∑€å
plt.legend()  # ŸÜŸÖÿß€åÿ¥ ÿ±ÿßŸáŸÜŸÖÿß€å ÿ±ŸÜ⁄Ø‚ÄåŸáÿß (label‚ÄåŸáÿß)
plt.grid(True)  # ŸÅÿπÿßŸÑ ⁄©ÿ±ÿØŸÜ ÿÆÿ∑Ÿàÿ∑ ÿ¥ÿ®⁄©Ÿá‚Äåÿß€å ÿØÿ± ŸÜŸÖŸàÿØÿßÿ±
plt.show()  # ŸÜŸÖÿß€åÿ¥ ŸÜŸÖŸàÿØÿßÿ±
```
