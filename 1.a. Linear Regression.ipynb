📌 الگوریتم رگرسیون خطی (Linear Regression) چیه؟

تصور کن می‌خوای بدونی:

هرچی بیشتر درس بخونی، نمره‌ات چقدر بهتر میشه؟ 🎯

مثلاً چندتا مثال داری:

Capture.JPG

اگر این نقاط رو روی یک کاغذ رسم کنی، تقریباً روی یک خط می‌افتند.
کامپیوتر می‌آید بهترین خط ممکن را بین این نقطه‌ها پیدا می‌کند.
بعد می‌تواند حدس بزند:

اگر ۵ ساعت درس بخونی، تقریباً چند می‌گیری؟ 🎉

به این روش حدس‌زدن می‌گویند: رگرسیون خطی یعنی «یک خط پیدا کن که داده‌ها رو بهترین شکل ممکن توضیح بده.»

✨ خیلی ساده:

📈 ورودی: ساعت درس خواندن

📈 خروجی: نمره

📈 کامپیوتر: یک خط می‌کشه که به داده‌ها بیشترین شباهت رو داشته باشه.

بعد می‌گه: اگه ورودی جدید بدی (مثلاً ۵ ساعت)، روی اون خط نگاه می‌کنه و جواب رو حدس می‌زنه.

✅ یک مثال جذاب پایتون

حالا بیایم همون داستان درس‌خوندن و نمره رو با پایتون و یک نمودار خوشگل ببینیم! 🎨

</div>


```python
import numpy as np  # وارد کردن کتابخانه numpy برای کار با آرایه‌های عددی
import matplotlib.pyplot as plt  # وارد کردن کتابخانه matplotlib برای رسم نمودار
from sklearn.linear_model import LinearRegression  # وارد کردن مدل رگرسیون خطی از sklearn

# داده‌ها: ساعت مطالعه و نمره
hours = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)  # آرایه‌ای از ساعات مطالعه (1 تا 5)، تبدیل به آرایه دو بعدی
scores = np.array([50, 60, 70, 80, 90])  # آرایه‌ای از نمرات مربوط به هر ساعت مطالعه

# ساخت مدل و آموزش
model = LinearRegression()  # ساخت یک مدل رگرسیون خطی
model.fit(hours, scores)  # آموزش مدل با داده‌ها (ساعات مطالعه و نمرات)

# پیش‌بینی برای یک ساعت جدید (مثلا 6 ساعت)
new_hour = np.array([[6]])  # ورودی جدید برای پیش‌بینی (۶ ساعت مطالعه)
predicted_score = model.predict(new_hour)  # پیش‌بینی نمره برای ۶ ساعت مطالعه
print(f"If you study for 6 hours, your grade will be approximately: {predicted_score[0]:.1f}")  # نمایش نتیجه: اگر ۶ ساعت درس بخونی نمره‌ات حدوداً چنده

# رسم نمودار
plt.scatter(hours, scores, color='blue', label='Real Data')  # رسم داده‌های واقعی به صورت نقطه‌های آبی
plt.plot(hours, model.predict(hours), color='red', label='Learned line')  # رسم خط یادگرفته‌شده توسط مدل با رنگ قرمز
plt.scatter(new_hour, predicted_score, color='green', s=100, label='New guess')  # رسم پیش‌بینی جدید (نقطه سبز بزرگ)
plt.xlabel('Study hours')  # برچسب محور افقی: ساعت مطالعه
plt.ylabel('Score')  # برچسب محور عمودی: نمره
plt.title('Predicting scores with linear regression')  # عنوان نمودار: پیش‌بینی نمره با رگرسیون خطی
plt.legend()  # نمایش راهنمای رنگ‌ها (label‌ها)
plt.grid(True)  # فعال کردن خطوط شبکه‌ای در نمودار
plt.show()  # نمایش نمودار
```
