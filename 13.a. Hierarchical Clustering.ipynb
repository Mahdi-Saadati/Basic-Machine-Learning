Agglomerative (ÿ™ÿ¨ŸÖ€åÿπ€å) ‚Äì ÿßÿ≤ Ÿæÿß€å€åŸÜ ÿ®Ÿá ÿ®ÿßŸÑÿß

ÿ®ÿπÿØ ŸÜÿ≤ÿØ€å⁄©‚Äåÿ™ÿ±€åŸÜ ÿ®⁄ÜŸá‚ÄåŸáÿß ÿ®ÿß ŸáŸÖ €å⁄©€å ŸÖ€å‚Äåÿ¥ŸÜ.

ŸáŸÖ€åŸÜ‚Äåÿ∑Ÿàÿ± ÿßÿØÿßŸÖŸá Ÿæ€åÿØÿß ŸÖ€å‚Äå⁄©ŸÜŸá ÿ™ÿß ŸáŸÖŸá ÿ™Ÿà€å €å⁄© ⁄Øÿ±ŸàŸá ÿ®ÿ≤ÿ±⁄Ø ÿ®ÿ¥ŸÜ.

Divisive (ÿ™ŸÇÿ≥€åŸÖ€å) ‚Äì ÿßÿ≤ ÿ®ÿßŸÑÿß ÿ®Ÿá Ÿæÿß€å€åŸÜ

ÿßŸàŸÑ ŸáŸÖŸá ÿ®⁄ÜŸá‚ÄåŸáÿß ÿ™Ÿà€å €å⁄© ⁄Øÿ±ŸàŸá ÿ®ÿ≤ÿ±⁄Ø Ÿáÿ≥ÿ™ŸÜ.

ÿ®ÿπÿØ ⁄Øÿ±ŸàŸá ÿ®Ÿá ÿØŸà ŸÇÿ≥ŸÖÿ™ ÿ™ŸÇÿ≥€åŸÖ ŸÖ€åÿ¥Ÿá.

ÿØŸàÿ®ÿßÿ±Ÿá ÿ™ŸÇÿ≥€åŸÖ ÿßÿØÿßŸÖŸá Ÿæ€åÿØÿß ŸÖ€å‚Äå⁄©ŸÜŸá ÿ™ÿß ⁄Øÿ±ŸàŸá‚ÄåŸáÿß€å ⁄©Ÿà⁄Ü⁄©ÿ™ÿ± ÿ≥ÿßÿÆÿ™Ÿá ÿ®ÿ¥ŸÜ.

üñº ÿØŸÜÿØÿ±Ÿà⁄Øÿ±ÿßŸÖ (Dendrogram)

ŸÜÿ™€åÿ¨Ÿá‚Äå€å Hierarchical Clustering ŸÖÿπŸÖŸàŸÑÿßŸã ÿ®Ÿá ÿµŸàÿ±ÿ™ €å⁄© ÿØÿ±ÿÆÿ™ (ÿ¥ÿ®€åŸá ÿ¥ÿ¨ÿ±Ÿá‚ÄåŸÜÿßŸÖŸá ÿÆÿßŸÜŸàÿßÿØ⁄Ø€å) ŸÜÿ¥ŸàŸÜ ÿØÿßÿØŸá ŸÖ€åÿ¥Ÿá.

Ÿáÿ± ÿ¥ÿßÿÆŸá‚Äå€å ÿØÿ±ÿÆÿ™ €åÿπŸÜ€å €å⁄© ÿÆŸàÿ¥Ÿá.

Ÿáÿ±⁄ÜŸá ÿßÿ±ÿ™ŸÅÿßÿπ ÿ¥ÿßÿÆŸá‚ÄåŸáÿß ÿ®€åÿ¥ÿ™ÿ± ÿ®ÿßÿ¥Ÿá €åÿπŸÜ€å ŸÅÿßÿµŸÑŸá‚Äå€å ÿ®€åŸÜ ⁄Øÿ±ŸàŸá‚ÄåŸáÿß ÿ®€åÿ¥ÿ™ÿ±Ÿá.

üêº ŸÖÿ´ÿßŸÑ Ÿæÿß€åÿ™ŸàŸÜ€å ÿ®ÿß Hierarchical Clustering

ÿ®€åÿß€å€åŸÖ €å⁄© ŸÖÿ´ÿßŸÑ ÿ®ÿß ÿØÿßÿØŸá‚ÄåŸáÿß€å ÿÆ€åÿßŸÑ€å ÿ®ÿ≤ŸÜ€åŸÖ:

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.cluster import AgglomerativeClustering

# 1. ÿ≥ÿßÿÆÿ™ ÿØÿßÿØŸá‚ÄåŸáÿß€å ÿÆ€åÿßŸÑ€å (ŸÖÿ´ŸÑÿßŸã ÿ®⁄ÜŸá‚ÄåŸáÿß ÿ®ÿß ŸÇÿØ Ÿà Ÿàÿ≤ŸÜ ŸÖÿÆÿ™ŸÑŸÅ)
X, _ = make_blobs(n_samples=20, centers=3, random_state=42)

# 2. ÿ≥ÿßÿÆÿ™ ÿØŸÜÿØÿ±Ÿà⁄Øÿ±ÿßŸÖ
linked = linkage(X, method='ward')

plt.figure(figsize=(8, 4))
dendrogram(linked)
plt.title("Hierarchical Clustring - Dendrogram")
plt.show()

# 3. ÿßÿ¨ÿ±ÿß€å Agglomerative Clustering ÿ®ÿ±ÿß€å 3 ÿÆŸàÿ¥Ÿá
cluster = AgglomerativeClustering(n_clusters=3, linkage='ward')
labels = cluster.fit_predict(X)

# 4. ŸÜŸÖÿß€åÿ¥ ÿÆŸàÿ¥Ÿá‚ÄåŸáÿß
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=50)
plt.title("Hierarchical Clustring - Agglomerative")
plt.show()
```
