Hierarchical Clustering (خوشه‌بندی سلسله‌مراتبی)

🎓 توضیح ساده Hierarchical Clustering

فرض کن توی مدرسه می‌خوای بچه‌ها رو بر اساس شباهتشون گروه‌بندی کنی.
این بار به جای اینکه از قبل بگی "چند گروه داریم"، کم‌کم گروه‌ها رو می‌سازی.

دو روش اصلی:

Agglomerative (تجمیعی) – از پایین به بالا

هر بچه اول خودش یه گروه مستقل هست.

بعد نزدیک‌ترین بچه‌ها با هم یکی می‌شن.

همین‌طور ادامه پیدا می‌کنه تا همه توی یک گروه بزرگ بشن.

Divisive (تقسیمی) – از بالا به پایین

اول همه بچه‌ها توی یک گروه بزرگ هستن.

بعد گروه به دو قسمت تقسیم میشه.

دوباره تقسیم ادامه پیدا می‌کنه تا گروه‌های کوچکتر ساخته بشن.

🖼 دندروگرام (Dendrogram)

نتیجه‌ی Hierarchical Clustering معمولاً به صورت یک درخت (شبیه شجره‌نامه خانوادگی) نشون داده میشه.

هر شاخه‌ی درخت یعنی یک خوشه.

هرچه ارتفاع شاخه‌ها بیشتر باشه یعنی فاصله‌ی بین گروه‌ها بیشتره.

🐼 مثال پایتونی با Hierarchical Clustering

بیاییم یک مثال با داده‌های خیالی بزنیم:

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.cluster import AgglomerativeClustering

# 1. ساخت داده‌های خیالی (مثلاً بچه‌ها با قد و وزن مختلف)
X, _ = make_blobs(n_samples=20, centers=3, random_state=42)

# 2. ساخت دندروگرام
linked = linkage(X, method='ward')

plt.figure(figsize=(8, 4))
dendrogram(linked)
plt.title("Hierarchical Clustring - Dendrogram")
plt.show()

# 3. اجرای Agglomerative Clustering برای 3 خوشه
cluster = AgglomerativeClustering(n_clusters=3, linkage='ward')
labels = cluster.fit_predict(X)

# 4. نمایش خوشه‌ها
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=50)
plt.title("Hierarchical Clustring - Agglomerative")
plt.show()
```

✨ خروجی چه می‌شود؟

در دندروگرام: می‌بینیم که داده‌ها مرحله‌به‌مرحله به هم متصل می‌شوند.

در نمودار خوشه‌بندی: داده‌ها به سه رنگ مختلف (سه گروه) تقسیم شده‌اند.

🔑 روش‌های محاسبه فاصله بین خوشه‌ها در این الگوریتم:

Single Linkage → فاصله نزدیک‌ترین نقاط

Complete Linkage → فاصله دورترین نقاط

Average Linkage → میانگین فاصله‌ها

Ward’s Method → کمترین افزایش واریانس (خیلی پرکاربرد)
