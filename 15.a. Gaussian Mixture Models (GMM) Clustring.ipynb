Gaussian Mixture Models (GMM)

ایده ساده: فرض می‌کند که داده‌ها از ترکیب چند توزیع نرمال (گاوسی) تولید شده‌اند.

بر خلاف K-Means: هر داده می‌تواند با احتمال‌های مختلف به چند خوشه تعلق داشته باشد (نرم‌تر است).

مناسب برای: زمانی که خوشه‌ها همپوشانی دارند.

🎓 توضیح ساده GMM

فرض کن توی کلاس درس، قد دانش‌آموزها رو اندازه‌گیری می‌کنیم.

یه گروه قدکوتاه هستن.

یه گروه قدبلند.

یه گروه هم متوسط.

حالا می‌خوایم این گروه‌ها رو پیدا کنیم.

الگوریتم GMM فرض می‌کنه داده‌ها از ترکیب چند توزیع نرمال (Gaussian) ساخته شده‌ان.
هر توزیع نرمال مثل یک "کوه زنگوله‌ای شکل" است که روی داده‌ها قرار می‌گیرد.

فرقش با K-Means:

در K-Means، هر داده دقیقاً به یک خوشه تعلق دارد. (قطعی ✅❌)

در GMM، هر داده می‌تواند با احتمال‌های مختلف عضو چند خوشه باشد. (نرم‌تر → احتمالی 🎲)

مثلاً:
یک دانش‌آموز 70٪ احتمال دارد در گروه "قدبلندها" باشد و 30٪ در گروه "متوسط‌ها".

🔑 نکته مهم

GMM به کمک الگوریتم EM (Expectation-Maximization) یاد می‌گیرد:

حدس می‌زند داده‌ها به کدام توزیع تعلق دارند.

پارامترهای توزیع‌ها را دوباره به‌روزرسانی می‌کند.

این کار تکرار می‌شود تا بهترین مدل پیدا شود.

🐼 مثال پایتونی با GMM

بیاییم داده‌ی خیالی بسازیم و GMM را روی آن اجرا کنیم:

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.mixture import GaussianMixture

# 1. ساخت داده‌های خیالی (مثلاً قد و وزن دانش‌آموزها)
X, _ = make_blobs(n_samples=300, centers=3, cluster_std=0.8, random_state=42)

# 2. اجرای GMM با 3 خوشه
gmm = GaussianMixture(n_components=3, random_state=42)
gmm.fit(X)
labels = gmm.predict(X)  # برچسب خوشه‌ها

# 3. رسم داده‌ها و مراکز
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=30)
plt.scatter(gmm.means_[:, 0], gmm.means_[:, 1], c='red', marker='X', s=200, label='Cluster Centers')
plt.title(" Gaussian Mixture Models (GMM) Clustering")
plt.legend()
plt.show()
```

✨ خروجی چه می‌شود؟

داده‌ها به سه خوشه تقسیم می‌شوند.

هر خوشه شبیه یک بیضی (نه فقط دایره مثل K-Means).

مراکز خوشه‌ها با X قرمز مشخص شده‌اند.

در پشت صحنه، هر نقطه یک احتمال دارد که به هر خوشه تعلق داشته باشد.
