ŸÅÿ±ÿ∂ ⁄©ŸÜ ÿ™Ÿà€å €å⁄© ÿ≤ŸÖ€åŸÜ ÿ®ÿßÿ≤€å Ÿæÿ± ÿßÿ≤ ÿ®⁄ÜŸá‚ÄåŸáÿß ÿß€åÿ≥ÿ™ÿßÿØ€å.

ÿßŸÑ⁄ØŸàÿ±€åÿ™ŸÖ Mean Shift ŸÖ€å‚Äå⁄ØŸá:

ÿß€åŸÜ ÿ≠ÿ±⁄©ÿ™ ÿßÿØÿßŸÖŸá Ÿæ€åÿØÿß ŸÖ€å‚Äå⁄©ŸÜŸá ÿ™ÿß ÿ®Ÿá "ŸÖÿ±⁄©ÿ≤ ÿ¥ŸÑŸàÿ∫€å" ÿ®ÿ±ÿ≥Ÿá.

üëâ ŸÜÿ™€åÿ¨Ÿá: ÿ®ÿØŸàŸÜ ÿß€åŸÜ⁄©Ÿá ÿßÿ≤ ŸÇÿ®ŸÑ ÿ®⁄Ø€å ⁄ÜŸÜÿØ ⁄Øÿ±ŸàŸá (K) Ÿàÿ¨ŸàÿØ ÿØÿßÿ±Ÿáÿå ÿÆŸàÿ¥Ÿá‚ÄåŸáÿß ÿ®Ÿá ÿ∑Ÿàÿ± ÿÆŸàÿØ⁄©ÿßÿ± ÿ≥ÿßÿÆÿ™Ÿá ŸÖ€å‚Äåÿ¥ŸÜ.

K-Means ÿ®ÿß€åÿØ ÿ™ÿπÿØÿßÿØ ÿÆŸàÿ¥Ÿá‚ÄåŸáÿß ÿ±Ÿà ÿßÿ≤ ŸÇÿ®ŸÑ ÿ®ÿØ€å.

Mean Shift ÿÆŸàÿØÿ¥ ÿ™ÿπÿØÿßÿØ ÿÆŸàÿ¥Ÿá‚ÄåŸáÿß ÿ±Ÿà Ÿæ€åÿØÿß ŸÖ€å‚Äå⁄©ŸÜŸá.

K-Means ŸÅŸÇÿ∑ ÿÆŸàÿ¥Ÿá‚ÄåŸáÿß€å ÿ™ŸÇÿ±€åÿ®ÿßŸã ⁄Øÿ±ÿØ ÿ±Ÿà ÿÆŸàÿ® Ÿæ€åÿØÿß ŸÖ€å‚Äå⁄©ŸÜŸá.

Mean Shift ŸÖ€å‚Äåÿ™ŸàŸÜŸá ÿ¥⁄©ŸÑ‚ÄåŸáÿß€å ŸÖÿÆÿ™ŸÑŸÅ ÿ±Ÿà ŸáŸÖ Ÿæ€åÿØÿß ⁄©ŸÜŸá.

üêº ŸÖÿ´ÿßŸÑ Ÿæÿß€åÿ™ŸàŸÜ€å ÿ®ÿß Mean Shift

ÿ®€åÿß€å€åŸÖ €å⁄© ŸÖÿ´ÿßŸÑ ÿ™ÿµŸà€åÿ±€å ÿ®ÿ≤ŸÜ€åŸÖ:

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import MeanShift

# 1. ÿ≥ÿßÿÆÿ™ ÿØÿßÿØŸá‚ÄåŸáÿß€å ÿÆ€åÿßŸÑ€å (ŸÖÿ´ŸÑ ⁄Øÿ±ŸàŸá ÿ®⁄ÜŸá‚ÄåŸáÿß ÿØÿ± ÿ≠€åÿßÿ∑)
X, _ = make_blobs(n_samples=300, centers=4, cluster_std=0.7, random_state=42)

# 2. ÿßÿ¨ÿ±ÿß€å ÿßŸÑ⁄ØŸàÿ±€åÿ™ŸÖ Mean Shift
ms = MeanShift()
ms.fit(X)
labels = ms.labels_         # ÿ®ÿ±⁄Üÿ≥ÿ® ÿÆŸàÿ¥Ÿá‚ÄåŸáÿß
centers = ms.cluster_centers_  # ŸÖÿ±ÿß⁄©ÿ≤ ÿÆŸàÿ¥Ÿá

# 3. ÿ±ÿ≥ŸÖ ŸÜŸÖŸàÿØÿßÿ±
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=30)
plt.scatter(centers[:, 0], centers[:, 1], c='red', marker='X', s=200, label='Cluster centers')
plt.title("Mean Shift clustring")
plt.legend()
plt.show()
```
