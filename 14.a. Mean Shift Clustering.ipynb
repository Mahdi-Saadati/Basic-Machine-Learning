 الگوریتم Mean Shift Clustering

 ایده ساده: داده‌ها به سمت ناحیه‌هایی با تراکم بالا «حرکت» می‌کنند تا خوشه‌ها تشکیل شوند.

نیاز به تعیین تعداد خوشه‌ها نیست.

مشابه K-Means اما با رویکرد متفاوت.

مناسب برای: پیدا کردن تعداد نامشخص خوشه‌ها با فرم‌های مختلف.

🎓 توضیح ساده Mean Shift

فرض کن توی یک زمین بازی پر از بچه‌ها ایستادی.

بعضی جاها بچه‌ها زیاد جمع شدن (مثلاً گوشه‌ی زمین فوتبال یا کنار تاب).

بعضی جاها فقط یکی‌دو نفر ایستادن.

الگوریتم Mean Shift می‌گه:

هر نقطه (بچه) شروع می‌کنه به حرکت کردن به سمت جایی که بچه‌های بیشتری جمع شدن.

این حرکت ادامه پیدا می‌کنه تا به "مرکز شلوغی" برسه.

هر کسی که به یک مرکز رسیده باشه، عضو همون خوشه حساب میشه.

👉 نتیجه: بدون اینکه از قبل بگی چند گروه (K) وجود داره، خوشه‌ها به طور خودکار ساخته می‌شن.

🔑 فرق Mean Shift با K-Means

K-Means باید تعداد خوشه‌ها رو از قبل بدی.

Mean Shift خودش تعداد خوشه‌ها رو پیدا می‌کنه.

K-Means فقط خوشه‌های تقریباً گرد رو خوب پیدا می‌کنه.

Mean Shift می‌تونه شکل‌های مختلف رو هم پیدا کنه.

🐼 مثال پایتونی با Mean Shift

بیاییم یک مثال تصویری بزنیم:

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import MeanShift

# 1. ساخت داده‌های خیالی (مثل گروه بچه‌ها در حیاط)
X, _ = make_blobs(n_samples=300, centers=4, cluster_std=0.7, random_state=42)

# 2. اجرای الگوریتم Mean Shift
ms = MeanShift()
ms.fit(X)
labels = ms.labels_         # برچسب خوشه‌ها
centers = ms.cluster_centers_  # مراکز خوشه

# 3. رسم نمودار
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=30)
plt.scatter(centers[:, 0], centers[:, 1], c='red', marker='X', s=200, label='Cluster centers')
plt.title("Mean Shift clustring")
plt.legend()
plt.show()
```

داده‌ها (بچه‌ها) با رنگ‌های مختلف به خوشه‌های جدا تقسیم می‌شوند.

علامت X قرمز نشان‌دهنده‌ی مرکز هر خوشه است.

تعداد خوشه‌ها به طور خودکار انتخاب شده (لازم نبود بگیم).
