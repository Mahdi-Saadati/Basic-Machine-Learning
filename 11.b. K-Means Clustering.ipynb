ÿ®€åÿß€å€åÿØ ÿßŸÑ⁄ØŸàÿ±€åÿ™ŸÖ K-Means ÿ±ÿß ÿ±Ÿà€å €å⁄© ÿØ€åÿ™ÿßÿ≥ÿ™ ŸàÿßŸÇÿπ€å Ÿà ÿ≥ÿßÿØŸá (ŸÖÿ´ŸÑ ÿØ€åÿ™ÿßÿ≥ÿ™ ŸÖÿπÿ±ŸàŸÅ Iris) ÿ®ÿß Ÿæÿß€åÿ™ŸàŸÜ Ÿæ€åÿßÿØŸá‚Äåÿ≥ÿßÿ≤€å ⁄©ŸÜ€åŸÖ.

üìå ÿ™Ÿàÿ∂€åÿ≠ ÿ≥ÿßÿØŸá

ÿßŸÑ⁄ØŸàÿ±€åÿ™ŸÖ K-Means ÿØÿßÿØŸá‚ÄåŸáÿß ÿ±ÿß ÿ®Ÿá ⁄ÜŸÜÿØ ⁄Øÿ±ŸàŸá (ÿÆŸàÿ¥Ÿá) ÿ™ŸÇÿ≥€åŸÖ ŸÖ€å‚Äå⁄©ŸÜÿØ. ŸÖÿ´ŸÑÿßŸã ÿß⁄Øÿ± ÿ®⁄ØŸà€å€åŸÖ k=3ÿå ÿØÿßÿØŸá‚ÄåŸáÿß ÿ±ÿß ÿ®Ÿá 3 ÿØÿ≥ÿ™Ÿá ÿ™ŸÇÿ≥€åŸÖ ŸÖ€å‚Äå⁄©ŸÜÿØ ⁄©Ÿá ÿØÿßÿØŸá‚ÄåŸáÿß€å ÿ¥ÿ®€åŸá ŸáŸÖ ÿØÿ± €å⁄© ÿØÿ≥ÿ™Ÿá ŸÇÿ±ÿßÿ± ⁄Ø€åÿ±ŸÜÿØ.

```python
# ⁄©ÿ™ÿßÿ®ÿÆÿßŸÜŸá‚ÄåŸáÿß
import matplotlib.pyplot as plt      # ÿ®ÿ±ÿß€å ÿ™ÿ±ÿ≥€åŸÖ ŸÜŸÖŸàÿØÿßÿ±Ÿáÿß
from sklearn.datasets import load_iris  # ÿ®ÿ±ÿß€å ÿ®ÿßÿ±⁄Øÿ∞ÿßÿ±€å ÿØ€åÿ™ÿßÿ≥ÿ™ ŸÖÿπÿ±ŸàŸÅ Iris
from sklearn.cluster import KMeans      # ÿßŸÑ⁄ØŸàÿ±€åÿ™ŸÖ KMeans ÿ®ÿ±ÿß€å ÿÆŸàÿ¥Ÿá‚Äåÿ®ŸÜÿØ€å
import pandas as pd                     # ÿ®ÿ±ÿß€å ⁄©ÿßÿ± ÿ®ÿß ÿØ€åÿ™ÿßŸÅÿ±€åŸÖ Ÿà ŸÜŸÖÿß€åÿ¥ ÿØÿßÿØŸá‚ÄåŸáÿß

# ÿ®ÿßÿ±⁄Øÿ∞ÿßÿ±€å ÿØ€åÿ™ÿßÿ≥ÿ™ Iris
iris = load_iris()     # ÿØ€åÿ™ÿßÿ≥ÿ™ ÿ¥ÿßŸÖŸÑ 150 ŸÜŸÖŸàŸÜŸá ⁄ØŸÑÿå ÿ®ÿß 4 Ÿà€å⁄ò⁄Ø€å (ÿ∑ŸàŸÑ/ÿπÿ±ÿ∂ ⁄©ÿßÿ≥ÿ®ÿ±⁄Ø Ÿà ⁄ØŸÑÿ®ÿ±⁄Ø)
X = iris.data          # ŸÅŸÇÿ∑ Ÿà€å⁄ò⁄Ø€å‚ÄåŸáÿß (featureŸáÿß)ÿå ŸÑ€åÿ®ŸÑ ÿßÿµŸÑ€å ⁄©ŸÑÿßÿ≥‚ÄåŸáÿß ÿ±ÿß ŸÜŸÖ€å‚Äå⁄Ø€åÿ±€åŸÖ

# ÿ™ÿπÿ±€åŸÅ ŸÖÿØŸÑ KMeans
kmeans = KMeans(n_clusters=3, random_state=42)  # ŸÖ€å‚ÄåÿÆŸàÿßŸá€åŸÖ ÿØÿßÿØŸá‚ÄåŸáÿß ÿ±ÿß ÿØÿ± 3 ÿÆŸàÿ¥Ÿá ⁄Øÿ±ŸàŸá‚Äåÿ®ŸÜÿØ€å ⁄©ŸÜ€åŸÖ
kmeans.fit(X)                                   # ÿßÿ¨ÿ±ÿß€å ÿßŸÑ⁄ØŸàÿ±€åÿ™ŸÖ KMeans ÿ±Ÿà€å ÿØÿßÿØŸá‚ÄåŸáÿß

# ⁄Øÿ±ŸÅÿ™ŸÜ ÿ®ÿ±⁄Üÿ≥ÿ® ÿÆŸàÿ¥Ÿá‚ÄåŸáÿß
labels = kmeans.labels_   # ÿ®ÿ±ÿß€å Ÿáÿ± ŸÜŸÖŸàŸÜŸáÿå ÿ¥ŸÖÿßÿ±Ÿá ÿÆŸàÿ¥Ÿá (€∞ÿå€±ÿå€≤) ŸÖÿ¥ÿÆÿµ ŸÖ€å‚Äåÿ¥ŸàÿØ

# ÿ≥ÿßÿÆÿ™ ÿØ€åÿ™ÿßŸÅÿ±€åŸÖ ÿ®ÿ±ÿß€å ŸÜŸÖÿß€åÿ¥ ÿØÿßÿØŸá‚ÄåŸáÿß ŸáŸÖÿ±ÿßŸá ÿ®ÿß ÿÆŸàÿ¥Ÿá
df = pd.DataFrame(X, columns=iris.feature_names)  # ÿ≥ÿßÿÆÿ™ ÿ¨ÿØŸàŸÑ ÿ®ÿß ÿ≥ÿ™ŸàŸÜ‚ÄåŸáÿß€å Ÿà€å⁄ò⁄Ø€å
df["Cluster"] = labels                           # ÿßÿ∂ÿßŸÅŸá ⁄©ÿ±ÿØŸÜ ÿ≥ÿ™ŸàŸÜ "ÿÆŸàÿ¥Ÿá" ÿ®Ÿá ÿ¨ÿØŸàŸÑ
print(df.head())                                 # ŸÜŸÖÿß€åÿ¥ €µ ÿ≥ÿ∑ÿ± ÿßŸàŸÑ ÿ¨ÿØŸàŸÑ

# ÿ™ÿ±ÿ≥€åŸÖ ŸÜÿ™ÿß€åÿ¨ ÿÆŸàÿ¥Ÿá‚Äåÿ®ŸÜÿØ€å ÿ®ÿß ÿßÿ≥ÿ™ŸÅÿßÿØŸá ÿßÿ≤ ÿØŸà Ÿà€å⁄ò⁄Ø€å ÿßŸàŸÑ (Sepal length Ÿà Sepal width)
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=50)  # ŸÜŸÖÿß€åÿ¥ ÿØÿßÿØŸá‚ÄåŸáÿß ÿ®ÿß ÿ±ŸÜ⁄Ø ÿÆŸàÿ¥Ÿá
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
            c='red', marker='X', s=200, label='Centers')       # ŸÜŸÖÿß€åÿ¥ ŸÖÿ±ÿß⁄©ÿ≤ ÿÆŸàÿ¥Ÿá ÿ®ÿß ÿπŸÑÿßŸÖÿ™ X ŸÇÿ±ŸÖÿ≤
plt.xlabel("Sepal length")   # ÿ®ÿ±⁄Üÿ≥ÿ® ŸÖÿ≠Ÿàÿ± X
plt.ylabel("Sepal width")    # ÿ®ÿ±⁄Üÿ≥ÿ® ŸÖÿ≠Ÿàÿ± Y
plt.legend()                 # ŸÜŸÖÿß€åÿ¥ ÿ±ÿßŸáŸÜŸÖÿß€å ŸÜŸÖŸàÿØÿßÿ±
plt.title("K-Means Clustering on Iris Dataset")  # ÿπŸÜŸàÿßŸÜ ŸÜŸÖŸàÿØÿßÿ±
plt.show()                   # ŸÜŸÖÿß€åÿ¥ ŸÜŸÖŸàÿØÿßÿ± ŸÜŸáÿß€å€å
```
