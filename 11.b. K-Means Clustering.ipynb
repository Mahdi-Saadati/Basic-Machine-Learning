بیایید الگوریتم K-Means را روی یک دیتاست واقعی و ساده (مثل دیتاست معروف Iris) با پایتون پیاده‌سازی کنیم.

📌 توضیح ساده

الگوریتم K-Means داده‌ها را به چند گروه (خوشه) تقسیم می‌کند. مثلاً اگر بگوییم k=3، داده‌ها را به 3 دسته تقسیم می‌کند که داده‌های شبیه هم در یک دسته قرار گیرند.

```python
# کتابخانه‌ها
import matplotlib.pyplot as plt      # برای ترسیم نمودارها
from sklearn.datasets import load_iris  # برای بارگذاری دیتاست معروف Iris
from sklearn.cluster import KMeans      # الگوریتم KMeans برای خوشه‌بندی
import pandas as pd                     # برای کار با دیتافریم و نمایش داده‌ها

# بارگذاری دیتاست Iris
iris = load_iris()     # دیتاست شامل 150 نمونه گل، با 4 ویژگی (طول/عرض کاسبرگ و گلبرگ)
X = iris.data          # فقط ویژگی‌ها (featureها)، لیبل اصلی کلاس‌ها را نمی‌گیریم

# تعریف مدل KMeans
kmeans = KMeans(n_clusters=3, random_state=42)  # می‌خواهیم داده‌ها را در 3 خوشه گروه‌بندی کنیم
kmeans.fit(X)                                   # اجرای الگوریتم KMeans روی داده‌ها

# گرفتن برچسب خوشه‌ها
labels = kmeans.labels_   # برای هر نمونه، شماره خوشه (۰،۱،۲) مشخص می‌شود

# ساخت دیتافریم برای نمایش داده‌ها همراه با خوشه
df = pd.DataFrame(X, columns=iris.feature_names)  # ساخت جدول با ستون‌های ویژگی
df["Cluster"] = labels                           # اضافه کردن ستون "خوشه" به جدول
print(df.head())                                 # نمایش ۵ سطر اول جدول

# ترسیم نتایج خوشه‌بندی با استفاده از دو ویژگی اول (Sepal length و Sepal width)
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=50)  # نمایش داده‌ها با رنگ خوشه
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
            c='red', marker='X', s=200, label='Centers')       # نمایش مراکز خوشه با علامت X قرمز
plt.xlabel("Sepal length")   # برچسب محور X
plt.ylabel("Sepal width")    # برچسب محور Y
plt.legend()                 # نمایش راهنمای نمودار
plt.title("K-Means Clustering on Iris Dataset")  # عنوان نمودار
plt.show()                   # نمایش نمودار نهایی
```
