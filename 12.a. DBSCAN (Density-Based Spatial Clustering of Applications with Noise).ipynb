 الگوریتم DBSCAN (Density-Based Spatial Clustering of Applications with Noise)

🎓 توضیح ساده DBSCAN

تصور کن توی حیاط مدرسه بچه‌ها دارن بازی می‌کنن:

بعضی‌ها گروهی دور هم جمع شدن (مثلاً ۵–۶ نفر نزدیک هم).

بعضی‌ها تنها ایستادن یا فاصله‌شون از بقیه زیاده.

حالا الگوریتم DBSCAN می‌گه:

هر وقت چند نفر خیلی به هم نزدیک باشن → اون‌ها رو یک خوشه در نظر می‌گیرم.

بچه‌هایی که خیلی دور ایستادن و همسایه ندارن → به عنوان نویز (outlier) در نظر می‌گیرم.

نکته‌ی مهم:

برخلاف K-Means:

لازم نیست از قبل بگی چند خوشه می‌خوای.

خوشه‌ها می‌تونن شکل‌های عجیب داشته باشن (دایره‌ای، خطی، یا نامنظم).

ایده ساده: خوشه‌ها مناطقی هستند که داده‌ها در آن‌ها متراکم هستند.

ویژگی جالب: نقاط پرت (outliers) را هم شناسایی می‌کند.

چگونه؟

برای هر نقطه، تعداد همسایه‌های نزدیک را بررسی می‌کند.

اگر زیاد باشند → نقطه در خوشه قرار می‌گیرد.

اگر کم باشند → نقطه به عنوان نویز یا مرزی در نظر گرفته می‌شود.

مناسب برای: داده‌های پراکنده یا با شکل نامنظم (نه فقط دایره‌ای).

🐼 مثال پایتونی DBSCAN

بیاییم یک مثال تصویری با داده‌های خیالی بزنیم:

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN

# 1. ساخت داده‌های خیالی (شکل نیم‌دایره‌ای مثل ماه)
X, _ = make_moons(n_samples=200, noise=0.05, random_state=42)

# 2. اجرای الگوریتم DBSCAN
dbscan = DBSCAN(eps=0.2, min_samples=5)
labels = dbscan.fit_predict(X)

# 3. رسم داده‌ها
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=30)
plt.title("DBSCAN Algorithm")
plt.show()
```
