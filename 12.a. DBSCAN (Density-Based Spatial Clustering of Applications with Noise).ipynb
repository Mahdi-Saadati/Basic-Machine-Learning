ŸÖŸÜÿßÿ≥ÿ® ÿ®ÿ±ÿß€å: ÿØÿßÿØŸá‚ÄåŸáÿß€å Ÿæÿ±ÿß⁄©ŸÜÿØŸá €åÿß ÿ®ÿß ÿ¥⁄©ŸÑ ŸÜÿßŸÖŸÜÿ∏ŸÖ (ŸÜŸá ŸÅŸÇÿ∑ ÿØÿß€åÿ±Ÿá‚Äåÿß€å).

üêº ŸÖÿ´ÿßŸÑ Ÿæÿß€åÿ™ŸàŸÜ€å DBSCAN

ÿ®€åÿß€å€åŸÖ €å⁄© ŸÖÿ´ÿßŸÑ ÿ™ÿµŸà€åÿ±€å ÿ®ÿß ÿØÿßÿØŸá‚ÄåŸáÿß€å ÿÆ€åÿßŸÑ€å ÿ®ÿ≤ŸÜ€åŸÖ:

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN

# 1. ÿ≥ÿßÿÆÿ™ ÿØÿßÿØŸá‚ÄåŸáÿß€å ÿÆ€åÿßŸÑ€å (ÿ¥⁄©ŸÑ ŸÜ€åŸÖ‚ÄåÿØÿß€åÿ±Ÿá‚Äåÿß€å ŸÖÿ´ŸÑ ŸÖÿßŸá)
X, _ = make_moons(n_samples=200, noise=0.05, random_state=42)

# 2. ÿßÿ¨ÿ±ÿß€å ÿßŸÑ⁄ØŸàÿ±€åÿ™ŸÖ DBSCAN
dbscan = DBSCAN(eps=0.2, min_samples=5)
labels = dbscan.fit_predict(X)

# 3. ÿ±ÿ≥ŸÖ ÿØÿßÿØŸá‚ÄåŸáÿß
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=30)
plt.title("DBSCAN Algorithm")
plt.show()
```
