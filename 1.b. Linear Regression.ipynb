⁄Ü⁄ØŸàŸÜŸá ÿ±Ÿà€å ÿ≥ÿ∑ÿ≠ ŸáŸÖŸà⁄ØŸÑŸàÿ®€åŸÜ A1C (€å⁄© ÿπÿØÿØ ⁄©Ÿá ⁄©ŸÜÿ™ÿ±ŸÑ ÿØ€åÿßÿ®ÿ™ ÿ±ÿß ŸÜÿ¥ÿßŸÜ ŸÖ€å‚ÄåÿØŸáÿØ) ÿßÿ´ÿ± ŸÖ€å‚Äå⁄Øÿ∞ÿßÿ±ÿØ.

ŸÖÿß €å⁄© ÿ≥ÿ±€å ÿØÿßÿØŸá‚Äå€å ŸÅÿ±ÿ∂€å ÿØÿßÿ±€åŸÖ (ÿ®ÿ±ÿß€å ÿ¢ŸÖŸàÿ≤ÿ¥)ÿå
Ÿà ŸÖ€å‚ÄåÿÆŸàÿßŸá€åŸÖ ÿ®ÿß ÿ±⁄Øÿ±ÿ≥€åŸàŸÜ ÿÆÿ∑€å ÿÆÿ∑€å Ÿæ€åÿØÿß ⁄©ŸÜ€åŸÖ ⁄©Ÿá ÿßÿ≤ ÿ±Ÿà€å ŸÇŸÜÿØ ÿÆŸàŸÜÿå A1C ÿ±ÿß ÿ≠ÿØÿ≥ ÿ®ÿ≤ŸÜÿØ.

```python
import numpy as np  # Ÿàÿßÿ±ÿØ ⁄©ÿ±ÿØŸÜ ⁄©ÿ™ÿßÿ®ÿÆÿßŸÜŸá numpy ÿ®ÿ±ÿß€å ⁄©ÿßÿ± ÿ®ÿß ÿ¢ÿ±ÿß€åŸá‚ÄåŸáÿß Ÿà ŸÖÿ≠ÿßÿ≥ÿ®ÿßÿ™ ÿπÿØÿØ€å
import matplotlib.pyplot as plt  # Ÿàÿßÿ±ÿØ ⁄©ÿ±ÿØŸÜ ⁄©ÿ™ÿßÿ®ÿÆÿßŸÜŸá matplotlib ÿ®ÿ±ÿß€å ÿ±ÿ≥ŸÖ ŸÜŸÖŸàÿØÿßÿ±
from sklearn.linear_model import LinearRegression  # Ÿàÿßÿ±ÿØ ⁄©ÿ±ÿØŸÜ ŸÖÿØŸÑ ÿ±⁄Øÿ±ÿ≥€åŸàŸÜ ÿÆÿ∑€å ÿßÿ≤ ⁄©ÿ™ÿßÿ®ÿÆÿßŸÜŸá sklearn

# --------------------------
# ÿØÿßÿØŸá‚ÄåŸáÿß€å ŸÅÿ±ÿ∂€å
# ŸÇŸÜÿØ ÿÆŸàŸÜ ŸÜÿßÿ¥ÿ™ÿß (mg/dL)
X = np.array([80, 90, 100, 110, 120, 130, 140, 150, 160, 170]).reshape(-1, 1)  # ŸÑ€åÿ≥ÿ™€å ÿßÿ≤ ŸÖŸÇÿßÿØ€åÿ± ŸÇŸÜÿØ ÿÆŸàŸÜ ŸÜÿßÿ¥ÿ™ÿßÿå ÿ™ÿ®ÿØ€åŸÑ ÿ®Ÿá ÿ¢ÿ±ÿß€åŸá ÿØŸà ÿ®ÿπÿØ€å

# ÿ≥ÿ∑ÿ≠ A1C ÿ™ŸÇÿ±€åÿ®€å (%)
y = np.array([5.0, 5.2, 5.4, 5.7, 6.0, 6.3, 6.7, 7.0, 7.4, 7.8])  # ŸÖŸÇÿßÿØ€åÿ± A1C ŸÖÿ™ŸÜÿßÿ∏ÿ± ÿ®ÿß ŸÇŸÜÿØ ÿÆŸàŸÜ‚ÄåŸáÿß

# --------------------------
# ÿ≥ÿßÿÆÿ™ Ÿà ÿ¢ŸÖŸàÿ≤ÿ¥ ŸÖÿØŸÑ
model = LinearRegression()  # ÿ≥ÿßÿÆÿ™ €å⁄© ŸÖÿØŸÑ ÿ±⁄Øÿ±ÿ≥€åŸàŸÜ ÿÆÿ∑€å
model.fit(X, y)  # ÿ¢ŸÖŸàÿ≤ÿ¥ ŸÖÿØŸÑ ÿ®ÿß ÿØÿßÿØŸá‚ÄåŸáÿß€å X (ŸÇŸÜÿØ ÿÆŸàŸÜ) Ÿà y (ÿ≥ÿ∑ÿ≠ A1C)

# Ÿæ€åÿ¥‚Äåÿ®€åŸÜ€å A1C ÿ®ÿ±ÿß€å ŸáŸÖÿßŸÜ ÿØÿßÿØŸá‚ÄåŸáÿß
y_pred = model.predict(X)  # Ÿæ€åÿ¥‚Äåÿ®€åŸÜ€å ŸÖŸÇÿßÿØ€åÿ± A1C ÿ®ÿß ÿßÿ≥ÿ™ŸÅÿßÿØŸá ÿßÿ≤ ŸÖÿØŸÑ ÿ¢ŸÖŸàÿ≤ÿ¥‚ÄåÿØ€åÿØŸá

# ⁄ÜÿßŸæ ÿ¥€åÿ® Ÿà ÿπÿ±ÿ∂ ÿßÿ≤ ŸÖÿ®ÿØÿ£
print("a=Slope of the line (coefficient):", model.coef_[0])  # ŸÜŸÖÿß€åÿ¥ ÿ¥€åÿ® ÿÆÿ∑ (a ÿØÿ± y = ax + b)
print("b=Width from origin:", model.intercept_)  # ŸÜŸÖÿß€åÿ¥ ÿπÿ±ÿ∂ ÿßÿ≤ ŸÖÿ®ÿØÿ£ (b ÿØÿ± y = ax + b)

xtest  = np.array([[0],[2],[4]]).reshape(-1,1)  # ÿ≥ÿßÿÆÿ™ ÿ¢ÿ±ÿß€åŸá‚Äåÿß€å ÿßÿ≤ Ÿàÿ±ŸàÿØ€å‚ÄåŸáÿß€å ÿ¨ÿØ€åÿØ ÿ®ÿ±ÿß€å ÿ™ÿ≥ÿ™
print("x=",xtest)  # ŸÜŸÖÿß€åÿ¥ Ÿàÿ±ŸàÿØ€å‚ÄåŸáÿß
ypred = model.predict(xtest)  # Ÿæ€åÿ¥‚Äåÿ®€åŸÜ€å ÿÆÿ±Ÿàÿ¨€å ÿ®ÿ±ÿß€å Ÿàÿ±ŸàÿØ€å‚ÄåŸáÿß€å ÿ™ÿ≥ÿ™€å (y = ax + b)
print("y=ax+b=",ypred)  # ŸÜŸÖÿß€åÿ¥ ÿÆÿ±Ÿàÿ¨€å‚ÄåŸáÿß€å Ÿæ€åÿ¥‚Äåÿ®€åŸÜ€å‚Äåÿ¥ÿØŸá

# --------------------------
# ÿ±ÿ≥ŸÖ ŸÜŸÖŸàÿØÿßÿ±
plt.scatter(X, y, color='blue', label='Real data (patients)')  # ÿ±ÿ≥ŸÖ ÿØÿßÿØŸá‚ÄåŸáÿß€å ŸàÿßŸÇÿπ€å (ŸÜŸÇÿßÿ∑ ÿ¢ÿ®€å)
plt.plot(X, y_pred, color='red', label='Regression line (model)')  # ÿ±ÿ≥ŸÖ ÿÆÿ∑ Ÿæ€åÿ¥‚Äåÿ®€åŸÜ€å‚Äåÿ¥ÿØŸá ÿ™Ÿàÿ≥ÿ∑ ŸÖÿØŸÑ (ÿÆÿ∑ ŸÇÿ±ŸÖÿ≤)
plt.xlabel('Fasting blood sugar (mg/dL)')  # ÿπŸÜŸàÿßŸÜ ŸÖÿ≠Ÿàÿ± ÿßŸÅŸÇ€å: ŸÇŸÜÿØ ÿÆŸàŸÜ ŸÜÿßÿ¥ÿ™ÿß
plt.ylabel('A1C level (%)')  # ÿπŸÜŸàÿßŸÜ ŸÖÿ≠Ÿàÿ± ÿπŸÖŸàÿØ€å: ÿ≥ÿ∑ÿ≠ A1C
plt.title('The relationship between blood sugar and A1C in diabetic patients')  # ÿπŸÜŸàÿßŸÜ ŸÜŸÖŸàÿØÿßÿ±: ÿ±ÿßÿ®ÿ∑Ÿá ÿ®€åŸÜ ŸÇŸÜÿØ ÿÆŸàŸÜ Ÿà A1C
plt.legend()  # ŸÜŸÖÿß€åÿ¥ ÿ±ÿßŸáŸÜŸÖÿß€å ÿ±ŸÜ⁄Ø‚ÄåŸáÿß
plt.show()  # ŸÜŸÖÿß€åÿ¥ ŸÜŸÖŸàÿØÿßÿ±
```
