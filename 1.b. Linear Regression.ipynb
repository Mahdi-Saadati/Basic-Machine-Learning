🎯 ایده‌ی مثال

فرض کنید می‌خواهیم بررسی کنیم که میزان قند خون ناشتا (Fasting Blood Sugar) یک فرد
چگونه روی سطح هموگلوبین A1C (یک عدد که کنترل دیابت را نشان می‌دهد) اثر می‌گذارد.

ما یک سری داده‌ی فرضی داریم (برای آموزش)،
و می‌خواهیم با رگرسیون خطی خطی پیدا کنیم که از روی قند خون، A1C را حدس بزند.



</div>


```python
import numpy as np  # وارد کردن کتابخانه numpy برای کار با آرایه‌ها و محاسبات عددی
import matplotlib.pyplot as plt  # وارد کردن کتابخانه matplotlib برای رسم نمودار
from sklearn.linear_model import LinearRegression  # وارد کردن مدل رگرسیون خطی از کتابخانه sklearn

# --------------------------
# داده‌های فرضی
# قند خون ناشتا (mg/dL)
X = np.array([80, 90, 100, 110, 120, 130, 140, 150, 160, 170]).reshape(-1, 1)  # لیستی از مقادیر قند خون ناشتا، تبدیل به آرایه دو بعدی

# سطح A1C تقریبی (%)
y = np.array([5.0, 5.2, 5.4, 5.7, 6.0, 6.3, 6.7, 7.0, 7.4, 7.8])  # مقادیر A1C متناظر با قند خون‌ها

# --------------------------
# ساخت و آموزش مدل
model = LinearRegression()  # ساخت یک مدل رگرسیون خطی
model.fit(X, y)  # آموزش مدل با داده‌های X (قند خون) و y (سطح A1C)

# پیش‌بینی A1C برای همان داده‌ها
y_pred = model.predict(X)  # پیش‌بینی مقادیر A1C با استفاده از مدل آموزش‌دیده

# چاپ شیب و عرض از مبدأ
print("a=Slope of the line (coefficient):", model.coef_[0])  # نمایش شیب خط (a در y = ax + b)
print("b=Width from origin:", model.intercept_)  # نمایش عرض از مبدأ (b در y = ax + b)

xtest  = np.array([[0],[2],[4]]).reshape(-1,1)  # ساخت آرایه‌ای از ورودی‌های جدید برای تست
print("x=",xtest)  # نمایش ورودی‌ها
ypred = model.predict(xtest)  # پیش‌بینی خروجی برای ورودی‌های تستی (y = ax + b)
print("y=ax+b=",ypred)  # نمایش خروجی‌های پیش‌بینی‌شده

# --------------------------
# رسم نمودار
plt.scatter(X, y, color='blue', label='Real data (patients)')  # رسم داده‌های واقعی (نقاط آبی)
plt.plot(X, y_pred, color='red', label='Regression line (model)')  # رسم خط پیش‌بینی‌شده توسط مدل (خط قرمز)
plt.xlabel('Fasting blood sugar (mg/dL)')  # عنوان محور افقی: قند خون ناشتا
plt.ylabel('A1C level (%)')  # عنوان محور عمودی: سطح A1C
plt.title('The relationship between blood sugar and A1C in diabetic patients')  # عنوان نمودار: رابطه بین قند خون و A1C
plt.legend()  # نمایش راهنمای رنگ‌ها
plt.show()  # نمایش نمودار
```

🔎 چه اتفاقی می‌افتد؟

✅ نقاط آبی = داده‌های فرضی بیماران (قند خون و A1C آن‌ها)

✅ خط قرمز = خطی که مدل یاد گرفته تا بتواند A1C را از روی قند خون تخمین بزند.

✏️ توضیح ساده

این کد دارد به کامپیوتر یاد می‌دهد که بین قند خون و وضعیت دیابت (A1C) یک رابطه پیدا کند.
وقتی کامپیوتر این رابطه را پیدا کرد، می‌توانیم به آن بگوییم:
«اگر قند خون یک بیمار 145 باشد، حدوداً A1C او چقدر است؟»
و کامپیوتر با کمک این خط قرمز یک جواب تقریبی به ما می‌دهد.
