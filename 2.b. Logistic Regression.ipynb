```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression

# -----------------------------
# ÿØÿßÿØŸá‚ÄåŸáÿß
X = np.array([80, 90, 100, 110, 120, 130, 140, 150]).reshape(-1, 1)
y = np.array([0, 0, 0, 0, 1, 1, 1, 1])  # 0 = ÿ≥ÿßŸÑŸÖ, 1 = ÿØ€åÿßÿ®ÿ™

# -----------------------------
# ÿ≥ÿßÿÆÿ™ Ÿà ÿ¢ŸÖŸàÿ≤ÿ¥ ŸÖÿØŸÑ ŸÑÿ¨ÿ≥ÿ™€å⁄©
model = LogisticRegression()
model.fit(X, y)

# Ÿæ€åÿ¥‚Äåÿ®€åŸÜ€å ÿßÿ≠ÿ™ŸÖÿßŸÑ ÿØ€åÿßÿ®ÿ™ ÿ®ÿ±ÿß€å ŸáŸÖÿßŸÜ ÿØÿßÿØŸá‚ÄåŸáÿß
y_prob = model.predict_proba(X)[:,1]  # ŸÅŸÇÿ∑ ÿßÿ≠ÿ™ŸÖÿßŸÑ ⁄©ŸÑÿßÿ≥ 1 (ÿØ€åÿßÿ®ÿ™)
y_pred = model.predict(X)

# -----------------------------
# ⁄ÜÿßŸæ ŸÜÿ™ÿß€åÿ¨
print("Model prediction:", y_pred) #Ÿæ€åÿ¥‚Äåÿ®€åŸÜ€å ŸÖÿØŸÑ
print("Chance of diabetes:", y_prob) #ÿßÿ≠ÿ™ŸÖÿßŸÑ ÿØ€åÿßÿ®ÿ™:

# -----------------------------
# ÿ±ÿ≥ŸÖ ŸÜŸÖŸàÿØÿßÿ±
plt.scatter(X, y, color='blue', label='Real data') #ÿØÿßÿØŸá‚ÄåŸáÿß€å ŸàÿßŸÇÿπ€å
plt.plot(X, y_prob, color='red', label='Probability of diabetes (model)') #ÿßÿ≠ÿ™ŸÖÿßŸÑ ÿØ€åÿßÿ®ÿ™ (ŸÖÿØŸÑ)
plt.xlabel('Fasting blood sugar (mg/dL)') #ŸÇŸÜÿØ ÿÆŸàŸÜ ŸÜÿßÿ¥ÿ™ÿß (mg/dL)
plt.ylabel('Possibility of diabetes') #ÿßÿ≠ÿ™ŸÖÿßŸÑ ÿØ€åÿßÿ®ÿ™
plt.title('Logistic regression for diabetes prediction') #ÿ±⁄Øÿ±ÿ≥€åŸàŸÜ ŸÑÿ¨ÿ≥ÿ™€å⁄© ÿ®ÿ±ÿß€å Ÿæ€åÿ¥‚Äåÿ®€åŸÜ€å ÿØ€åÿßÿ®ÿ™
plt.legend()
plt.show()
```
