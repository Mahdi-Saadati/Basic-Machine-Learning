🧠 1. K-Means Clustering (خوشه‌بندی K-میانگین)

فرض کن توی یک مدرسه هستیم و می‌خواهیم بچه‌ها رو بر اساس قد و وزنشون به چند گروه تقسیم کنیم.

بعضی‌ها قدبلند و لاغر هستن

بعضی‌ها قدکوتاه و چاق هستن

بعضی‌ها هم بینابین

الگوریتم K-Means مثل یه معلم عمل می‌کنه:

اول حدس می‌زنه چند تا گروه (K) وجود داره. مثلا می‌گه "سه تا گروه".

سه تا نقطه تصادفی انتخاب می‌کنه و میگه: "شماها رهبر گروه هستید".

بچه‌ها میرن نزدیک‌ترین رهبر رو انتخاب می‌کنن. (هر داده به نزدیک‌ترین مرکز می‌چسبه)

حالا معلم جای رهبرها رو عوض می‌کنه تا دقیقا وسط بچه‌های گروهشون قرار بگیرن.

این کار تکرار میشه تا وقتی که رهبرها دیگه جاشون تغییر نکنن.

🎉 و اینطوری خوشه‌ها ساخته میشن!


ایده ساده: داده‌ها را به تعداد مشخصی از خوشه‌ها (K تا) تقسیم می‌کند.

چگونه؟:

چند نقطه تصادفی به عنوان مراکز خوشه انتخاب می‌شود.

هر داده به نزدیک‌ترین مرکز اختصاص می‌یابد.

مرکز خوشه‌ها دوباره محاسبه می‌شود.

مراحل ۲ و ۳ تا زمانی تکرار می‌شود که مراکز تغییر زیادی نکنند.

مثال ساده: فرض کنید می‌خواهیم مشتریان یک فروشگاه را بر اساس الگوی خریدشان گروه‌بندی کنیم.

🐼 مثال جذاب پایتونی با K-Means

در این مثال، می‌خوایم با پایتون یک سری داده‌ی خیالی بسازیم (مثلا بچه‌های مدرسه با قد و وزن مختلف) و اون‌ها رو خوشه‌بندی کنیم.

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

# 1. ساختن داده‌ی خیالی (مثل قد و وزن بچه‌ها)
X, _ = make_blobs(n_samples=200, centers=3, random_state=42)

# 2. اجرای الگوریتم K-Means برای 3 گروه
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)
labels = kmeans.labels_  # برچسب خوشه هر داده
centers = kmeans.cluster_centers_  # مرکز هر خوشه

# 3. رسم نمودار
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=30)
plt.scatter(centers[:, 0], centers[:, 1], c='red', marker='X', s=200, label='Cluster Centers')
plt.title("K-Means Algorithm")
plt.legend()
plt.show()
```
