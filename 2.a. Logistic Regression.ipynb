📌 رگرسیون لجستیک (Logistic Regression) یعنی چی؟

فرض کن می‌خواهیم بله/خیر یا قبول/رد را پیش‌بینی کنیم.
برای مثال:

🔸 اگر کسی چند ساعت درس بخواند، آیا قبول می‌شود یا رد؟

رگرسیون لجستیک مثل یک مدل می‌گوید:

اگر ساعت مطالعه کم باشد، احتمال قبول شدن پایین است.

هرچه ساعت مطالعه بیشتر شود، احتمال قبول شدن بالا می‌رود.
اما چون جواب فقط دو حالت دارد (قبول ✅ یا رد ❌)، این الگوریتم به جای یک خط صاف، یک منحنی S شکل می‌کشد که احتمال را از 0% تا 100% پیش‌بینی می‌کند.

✨ خیلی ساده:

✅ ورودی: تعداد ساعت مطالعه

✅ خروجی: قبول (1) یا رد (0)

✅ مدل: یک منحنی S شکل می‌کشد و می‌گوید برای هر ورودی احتمال قبول شدن چقدر است.

🎯 مثال جذاب پایتون

یک داستان بامزه:

ساعت‌های درس‌خواندن چند دانش‌آموز و اینکه قبول شدن یا نه.
مدل یاد می‌گیرد و بعد برای یک دانش‌آموز جدید پیش‌بینی می‌کند.




```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression

# داده‌های آموزشی: [ساعت مطالعه]
hours = np.array([1, 2, 3, 4, 5, 6, 7, 8]).reshape(-1, 1)
# برچسب‌ها: 0 = رد ❌ ، 1 = قبول ✅
passed = np.array([0, 0, 0, 0, 1, 1, 1, 1])

# ساخت مدل و آموزش
model = LogisticRegression()
model.fit(hours, passed)

# تست مدل با ساعت جدید
new_hour = np.array([[3.5]])
prob = model.predict_proba(new_hour)[0][1]  # احتمال قبول شدن
pred = model.predict(new_hour)[0]

print(f"if Study {new_hour[0][0]} hours:") #اگر ؟ ساعت درس بخوانی
print(f"Probability of acceptance: {prob*100:.1f}%") #احتمال قبول شدن
print("Final result:", "Accept" if pred==1 else "Reject") #نتیجه‌ی نهایی

# رسم نمودار
x_range = np.linspace(0, 9, 100).reshape(-1, 1)
y_prob = model.predict_proba(x_range)[:, 1]

plt.scatter(hours, passed, color='blue', label='Real data') #داده های واقعی
plt.plot(x_range, y_prob, color='red', linewidth=2, label='Logistic curve') #منحنی لجستیک
plt.axhline(0.5, color='gray', linestyle='--', label='limit 50%')
plt.xlabel('Study hour') #ساعت مطالعه
plt.ylabel('Probability of acceptance') #احتمال قبول شدن'
plt.title('Predicting acceptance or rejection with logistic regression') #پیش‌بینی قبول یا رد با رگرسیون لجستیک'
plt.legend()
plt.grid(True)
plt.show()
```
