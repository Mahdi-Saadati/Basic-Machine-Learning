üìå ÿ±⁄Øÿ±ÿ≥€åŸàŸÜ ŸÑÿ¨ÿ≥ÿ™€å⁄© (Logistic Regression) €åÿπŸÜ€å ⁄Ü€åÿü

ŸÅÿ±ÿ∂ ⁄©ŸÜ ŸÖ€å‚ÄåÿÆŸàÿßŸá€åŸÖ ÿ®ŸÑŸá/ÿÆ€åÿ± €åÿß ŸÇÿ®ŸàŸÑ/ÿ±ÿØ ÿ±ÿß Ÿæ€åÿ¥‚Äåÿ®€åŸÜ€å ⁄©ŸÜ€åŸÖ.
ÿ®ÿ±ÿß€å ŸÖÿ´ÿßŸÑ:

üî∏ ÿß⁄Øÿ± ⁄©ÿ≥€å ⁄ÜŸÜÿØ ÿ≥ÿßÿπÿ™ ÿØÿ±ÿ≥ ÿ®ÿÆŸàÿßŸÜÿØÿå ÿ¢€åÿß ŸÇÿ®ŸàŸÑ ŸÖ€å‚Äåÿ¥ŸàÿØ €åÿß ÿ±ÿØÿü

ÿ±⁄Øÿ±ÿ≥€åŸàŸÜ ŸÑÿ¨ÿ≥ÿ™€å⁄© ŸÖÿ´ŸÑ €å⁄© ŸÖÿØŸÑ ŸÖ€å‚Äå⁄ØŸà€åÿØ:

ÿß⁄Øÿ± ÿ≥ÿßÿπÿ™ ŸÖÿ∑ÿßŸÑÿπŸá ⁄©ŸÖ ÿ®ÿßÿ¥ÿØÿå ÿßÿ≠ÿ™ŸÖÿßŸÑ ŸÇÿ®ŸàŸÑ ÿ¥ÿØŸÜ Ÿæÿß€å€åŸÜ ÿßÿ≥ÿ™.

Ÿáÿ±⁄ÜŸá ÿ≥ÿßÿπÿ™ ŸÖÿ∑ÿßŸÑÿπŸá ÿ®€åÿ¥ÿ™ÿ± ÿ¥ŸàÿØÿå ÿßÿ≠ÿ™ŸÖÿßŸÑ ŸÇÿ®ŸàŸÑ ÿ¥ÿØŸÜ ÿ®ÿßŸÑÿß ŸÖ€å‚Äåÿ±ŸàÿØ.
ÿßŸÖÿß ⁄ÜŸàŸÜ ÿ¨Ÿàÿßÿ® ŸÅŸÇÿ∑ ÿØŸà ÿ≠ÿßŸÑÿ™ ÿØÿßÿ±ÿØ (ŸÇÿ®ŸàŸÑ ‚úÖ €åÿß ÿ±ÿØ ‚ùå)ÿå ÿß€åŸÜ ÿßŸÑ⁄ØŸàÿ±€åÿ™ŸÖ ÿ®Ÿá ÿ¨ÿß€å €å⁄© ÿÆÿ∑ ÿµÿßŸÅÿå €å⁄© ŸÖŸÜÿ≠ŸÜ€å S ÿ¥⁄©ŸÑ ŸÖ€å‚Äå⁄©ÿ¥ÿØ ⁄©Ÿá ÿßÿ≠ÿ™ŸÖÿßŸÑ ÿ±ÿß ÿßÿ≤ 0% ÿ™ÿß 100% Ÿæ€åÿ¥‚Äåÿ®€åŸÜ€å ŸÖ€å‚Äå⁄©ŸÜÿØ.

‚ú® ÿÆ€åŸÑ€å ÿ≥ÿßÿØŸá:

‚úÖ Ÿàÿ±ŸàÿØ€å: ÿ™ÿπÿØÿßÿØ ÿ≥ÿßÿπÿ™ ŸÖÿ∑ÿßŸÑÿπŸá

‚úÖ ÿÆÿ±Ÿàÿ¨€å: ŸÇÿ®ŸàŸÑ (1) €åÿß ÿ±ÿØ (0)

‚úÖ ŸÖÿØŸÑ: €å⁄© ŸÖŸÜÿ≠ŸÜ€å S ÿ¥⁄©ŸÑ ŸÖ€å‚Äå⁄©ÿ¥ÿØ Ÿà ŸÖ€å‚Äå⁄ØŸà€åÿØ ÿ®ÿ±ÿß€å Ÿáÿ± Ÿàÿ±ŸàÿØ€å ÿßÿ≠ÿ™ŸÖÿßŸÑ ŸÇÿ®ŸàŸÑ ÿ¥ÿØŸÜ ⁄ÜŸÇÿØÿ± ÿßÿ≥ÿ™.

üéØ ŸÖÿ´ÿßŸÑ ÿ¨ÿ∞ÿßÿ® Ÿæÿß€åÿ™ŸàŸÜ

€å⁄© ÿØÿßÿ≥ÿ™ÿßŸÜ ÿ®ÿßŸÖÿ≤Ÿá:

ÿ≥ÿßÿπÿ™‚ÄåŸáÿß€å ÿØÿ±ÿ≥‚ÄåÿÆŸàÿßŸÜÿØŸÜ ⁄ÜŸÜÿØ ÿØÿßŸÜÿ¥‚Äåÿ¢ŸÖŸàÿ≤ Ÿà ÿß€åŸÜ⁄©Ÿá ŸÇÿ®ŸàŸÑ ÿ¥ÿØŸÜ €åÿß ŸÜŸá.
ŸÖÿØŸÑ €åÿßÿØ ŸÖ€å‚Äå⁄Ø€åÿ±ÿØ Ÿà ÿ®ÿπÿØ ÿ®ÿ±ÿß€å €å⁄© ÿØÿßŸÜÿ¥‚Äåÿ¢ŸÖŸàÿ≤ ÿ¨ÿØ€åÿØ Ÿæ€åÿ¥‚Äåÿ®€åŸÜ€å ŸÖ€å‚Äå⁄©ŸÜÿØ.




```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression

# ÿØÿßÿØŸá‚ÄåŸáÿß€å ÿ¢ŸÖŸàÿ≤ÿ¥€å: [ÿ≥ÿßÿπÿ™ ŸÖÿ∑ÿßŸÑÿπŸá]
hours = np.array([1, 2, 3, 4, 5, 6, 7, 8]).reshape(-1, 1)
# ÿ®ÿ±⁄Üÿ≥ÿ®‚ÄåŸáÿß: 0 = ÿ±ÿØ ‚ùå ÿå 1 = ŸÇÿ®ŸàŸÑ ‚úÖ
passed = np.array([0, 0, 0, 0, 1, 1, 1, 1])

# ÿ≥ÿßÿÆÿ™ ŸÖÿØŸÑ Ÿà ÿ¢ŸÖŸàÿ≤ÿ¥
model = LogisticRegression()
model.fit(hours, passed)

# ÿ™ÿ≥ÿ™ ŸÖÿØŸÑ ÿ®ÿß ÿ≥ÿßÿπÿ™ ÿ¨ÿØ€åÿØ
new_hour = np.array([[3.5]])
prob = model.predict_proba(new_hour)[0][1]  # ÿßÿ≠ÿ™ŸÖÿßŸÑ ŸÇÿ®ŸàŸÑ ÿ¥ÿØŸÜ
pred = model.predict(new_hour)[0]

print(f"if Study {new_hour[0][0]} hours:") #ÿß⁄Øÿ± ÿü ÿ≥ÿßÿπÿ™ ÿØÿ±ÿ≥ ÿ®ÿÆŸàÿßŸÜ€å
print(f"Probability of acceptance: {prob*100:.1f}%") #ÿßÿ≠ÿ™ŸÖÿßŸÑ ŸÇÿ®ŸàŸÑ ÿ¥ÿØŸÜ
print("Final result:", "Accept" if pred==1 else "Reject") #ŸÜÿ™€åÿ¨Ÿá‚Äå€å ŸÜŸáÿß€å€å

# ÿ±ÿ≥ŸÖ ŸÜŸÖŸàÿØÿßÿ±
x_range = np.linspace(0, 9, 100).reshape(-1, 1)
y_prob = model.predict_proba(x_range)[:, 1]

plt.scatter(hours, passed, color='blue', label='Real data') #ÿØÿßÿØŸá Ÿáÿß€å ŸàÿßŸÇÿπ€å
plt.plot(x_range, y_prob, color='red', linewidth=2, label='Logistic curve') #ŸÖŸÜÿ≠ŸÜ€å ŸÑÿ¨ÿ≥ÿ™€å⁄©
plt.axhline(0.5, color='gray', linestyle='--', label='limit 50%')
plt.xlabel('Study hour') #ÿ≥ÿßÿπÿ™ ŸÖÿ∑ÿßŸÑÿπŸá
plt.ylabel('Probability of acceptance') #ÿßÿ≠ÿ™ŸÖÿßŸÑ ŸÇÿ®ŸàŸÑ ÿ¥ÿØŸÜ'
plt.title('Predicting acceptance or rejection with logistic regression') #Ÿæ€åÿ¥‚Äåÿ®€åŸÜ€å ŸÇÿ®ŸàŸÑ €åÿß ÿ±ÿØ ÿ®ÿß ÿ±⁄Øÿ±ÿ≥€åŸàŸÜ ŸÑÿ¨ÿ≥ÿ™€å⁄©'
plt.legend()
plt.grid(True)
plt.show()
```
